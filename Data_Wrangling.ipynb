The method used to deal with random number is called **random** and is **found in the numpy

These methods allow us to generate one random number at a time. <br/><br/>
What if we wanted to generate a **list of numbers, all lying within a given interval and generated randomly**.<br/> <br/>  Letâ€™s define a function that can generate a list of n random numbers lying between a and b.<br/>
- define a function, 
- wherein an empty list is created 
- and the randomly generated numbers are appended to the list.

```python
#method1

#defining a function

def randint_range(n,a,b):
 x=[]
 for i in range(n):               
  x.append(np.random.randint(a,b))
 return x

#calling a function

randint_range(10,2,1000)
```

```python
#method2 

#To generate three random numbers between 0 and 100, which are all multiples of 5,

import random
for i in range(3):
    print (random.randrange(0,100,5)) 
```

```python
#creating a simple dummy data frame that has a rich mix of both numerical and categorical variables.

#categorical random varibles

v1 = ['male','female']
v2 = ['poor','middle class','rich']
gender = []
socst = []

for x in range(1,101):
    gender.append(np.random.choice(v1))
    socst.append(np.random.choice(v2))    

#numerical random variables

age = 10 * np.random.randn(100) + 35
income =  1500 * np.random.randn(100) + 15000
ht =  30 * np.random.randn(100) + 155
wt =  20 * np.random.randn(100) + 60

#creating a dataframe

rdf = pd.DataFrame ({ 'Gender' : gender, 
                    'Soc_status': socst,
                    'Age': age,
                    'Income' : income,
                    'Height' : ht,
                    'Weight' : wt})
rdf.head()
```

##### Grouping

Grouping can be done over a categorical variable using the **groupby function**

```python
rdf.groupby('Gender')
```

```python
gro = rdf.groupby('Gender')
gro.groups
```

^|_ The numbers indicate the row numbers that belong to that particular group

```python
#This prints the name of the group followed by the entire data for this group.

gp = rdf.groupby('Gender')
for names,groups in gp: 
    print (names)
    print (groups)
```

```python
#A single group can be selected by writing the following:

gp.get_group('female')

#This would generate only the first of the two groups
```

A data frame can be grouped over more than one categorical variable as well.<br/>


```python
gp1 = rdf.groupby(['Gender','Soc_status'])
for names,groups in gp1:
    print(names)
    print(groups)
        
#6 groups will be executed.
```

The **aggregate method** is used to apply different functions to different columns.

```python
gp1.aggregate({ 'Income':np.sum,'Age':np.mean,'Height':np.std })
```

```python
gp1.aggregate({'Age':np.mean,'Height':lambda x:np.mean(x)/np.std(x)})
```

Rather than applying different functions to different columns, one can apply several
functions to all the columns at the same time.


```python
gp1.aggregate([np.sum, np.mean, np.std])
```

**'Transform method' : mathematically transform all the elements in a
numerical column**.


```python
zscore = lambda x: (x - x.mean()) / x.std()
gp1.transform(zscore)
```

The transform method comes in handy in a lot of situations.
> For example, it can be used
to fill the missing values with the mean of the non-missing values, as shown:


```python
f = lambda x: x.fillna(x.mean())
gp1.transform(f)
```
